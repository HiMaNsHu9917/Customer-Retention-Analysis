```python
df_clean = df[
    (~df['invoice'].astype(str).str.startswith(('C'), na=False)) & 
    (df['customer_id'].notnull()) &
    (df['quantity']>0) &
    (df['price']>0)
]
```

```python
df_clean.head()
```

```python
df_clean.info()
```

```python
df_clean['invoiceDate']=pd.to_datetime(df_clean['invoice_date'])
df_clean['customer ID']=df_clean['customer_id'].astype(int)
```

```python
df_clean.info()
```

```python
df_clean.drop(['description'],inplace=True,axis=1)
```

```python
df_clean.head()
```

```python
df_clean['total_price']=df_clean['price']*df_clean['quantity']
```

```python
# Creates a Period object (Year-Month)
df_clean['Month_Year'] = df_clean['invoice_date'].dt.to_period('M')
```

```python
df_clean.head()
```

```python
df_clean.columns=df_clean.columns.str.lower()
df_clean.columns=df_clean.columns.str.replace(' ','_')
```

```python
df_clean.head()
```

```python
df_clean.info()
```

```python
df_clean.rename(columns={'month_year': 'invoice_month',}, inplace=True)
```

```python
df_clean.info()
```

```python
df_clean.rename(columns={'invoicedate':'invoice_date'},inplace=True)
df_clean.drop(columns=['invoice_month'], inplace=True)
df_clean.to_csv('online_retail_II_full.csv', index=False)
```
