```python
import pandas as pd
from sqlalchemy import create_engine

df = pd.read_csv('data/online_retail_II_full.csv')

engine = create_engine('postgresql://postgres:admin123@127.0.0.1:5432/retail_project')

df.to_sql(
    name='online_retail',
    con=engine,
    if_exists='replace',
    index=False,
    chunksize=10000
)

print("Done! Table loaded successfully.")
```

```python
import psycopg2

conn = psycopg2.connect(
    host="127.0.0.1",
    port=5432,
    database="retail_project",
    user="postgres",
    password="admin123"
)
print("Connected!")
conn.close()
```
